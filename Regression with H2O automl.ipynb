```python
import h2o
from h2o.automl import H2OAutoML
import pandas as pd
```

```python

# Initialize H2O cluster
h2o.init()

# Load data
train = h2o.import_file('/content/drive/MyDrive/kag fellow/train_imputed_rf.csv')
test = h2o.import_file('/content/drive/MyDrive/kag fellow/test_final.csv')
```

### Infogram

```python
# Define features and target
features =train.drop(['price','id'],axis=1).columns
target = 'price'

# Run H2O AutoML
aml = H2OAutoML(seed=204,sort_metric='RMSE')
aml.train(x=features, y=target, training_frame=train)
```

```python
# View the AutoML Leaderboard
lb = aml.leaderboard
print(lb.head(rows=lb.nrows) )

# Get the best model
best_model = aml.leader

```

```python
# Make predictions on the test set
predictions = best_model.predict(test[features])
print(predictions)
# Get the best XGBoost model using default sort metric
xgb = aml.get_best_model(algorithm="xgboost",criterion='RMSE')
predictions = xgb.predict(test[features])
print(predictions)
```

```python
# View the parameters for the XGBoost model selected above
xgb.params.keys()
# Inspect individual parameter values
xgb.params['ntrees']
```

```python

# Evaluate model performance on training data (or use a separate validation set)
performance = best_model.model_performance(train)
print(performance)
```

```python

# Evaluate model performance on training data (or use a separate validation set)
performance = xgb.model_performance(train)
print(performance)
```

```python
predictions_df = predictions.as_data_frame()  # Convert predictions to a pandas DataFrame

# Add the predictions to the test DataFrame
test_df = test.as_data_frame()  # Convert test H2OFrame to pandas DataFrame
test_df['predicted_price'] = predictions_df['predict']

# Print the first few rows of the test DataFrame with predictions
print(test_df[['id', 'predicted_price']].head())

# Load the sample submission
sub = pd.read_csv('/content/drive/MyDrive/kag fellow/sample_submission.csv')

# Merge the submission template with the test DataFrame to create the final submission
submission = pd.merge(sub[['id']], test_df[['id', 'predicted_price']], on='id', how='left')

# Save the submission to a CSV file
submission.to_csv('/content/drive/MyDrive/kag fellow/new_h2o6.csv', index=False)
```
